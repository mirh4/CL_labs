**Задача 2**. Дан текстовый файл. Найти слово, встречающееся наиболее часто, и вывести количество его вхождений.

```python

with open('NNN.txt', encoding = 'utf-8') as f:
    st2 = f.read()
f.close()

words2 = re.split(r'\s', st2)

#words2 = re.split(r'[,.!?;\s]\s*', st2)

slovar2 = {}

for n in words2:
    if n in slovar2:
        slovar2[n] += 1
    else:
        slovar2[n] = 1

print(slovar2)  

import operator
val2 = []
for n in slovar2.values():
    val2.append(n)

print(val2)
    

new_ma_val = max(slovar2.items(), key=operator.itemgetter(1))[0]
print(' ')
print('Самое частое слово: ',(new_ma_val), '=', max(val2) )


```

```python

with open('input.txt', encoding = 'utf-8') as f:
    st3 = f.read()

f.close()

st3 = st3.lower()
words3 = re.sub(r'[\d]','', st3)
words3 = re.sub(r'<br>','',words3)
words3 = re.split(r'\W+',words3)
words3.pop(0)
num_words3 = len(words3)
words3.pop(num_words3 - 1)


#words3 = re.split(r'[br\W+\s\d\b]\s*', st3)
print(words3)
words3.sort()
#print(words3)
print('')
print('Частотный словарь:')

slovar3 = {}

for n in words3:
    if n in slovar3:
        slovar3[n] += 1
    else:
        slovar3[n] = 1

for x,y in slovar3.items():
    print(x, '=',y)
```
